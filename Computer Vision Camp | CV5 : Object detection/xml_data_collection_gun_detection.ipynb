```python
import pandas as pd
import numpy as np
import xml.etree.cElementTree as et
```

```python
jpg_file_paths = []
image_file_names = []
with open(r'F:\Eduard\Autodezvoltare\CV camp\CVC5\Gun_detection_xml\jpg_file_paths.txt') as f:
    for i in range(120):
        jpg_file_paths.append(f.readline())
for i in range(120):
    jpg_file_paths[i] = jpg_file_paths[i][1:-2]
for i in range(120):
    image_file_names.append(jpg_file_paths[i][62:])
image_file_names
```

```python
xml_file_paths = []
with open(r'F:\Eduard\Autodezvoltare\CV camp\CVC5\Gun_detection_xml\xml_file_paths.txt') as f:
    for i in range(120):
        xml_file_paths.append(f.readline())
for i in range(120):
    xml_file_paths[i] = xml_file_paths[i][1:-2]
xml_file_paths
```

```python
xmin_values = []
xmax_values = []
ymin_values = []
ymax_values = []
for i in range(120):
    
    tree=et.parse(xml_file_paths[i])
    root=tree.getroot()
    for item in root.iter('xmin'):
        xmin_values.append(int(item.text))
    for item in root.iter('xmax'):
        xmax_values.append(int(item.text))
    for item in root.iter('ymin'):
        ymin_values.append(int(item.text))
    for item in root.iter('ymax'):
        ymax_values.append(int(item.text))
```

```python
dictionary = {}
dictionary['image_path'] = image_file_names
dictionary['x_min'] = xmin_values
dictionary['y_min'] = ymin_values
dictionary['x_max'] = xmax_values
dictionary['y_max'] = ymax_values
```

```python
df = pd.DataFrame(dictionary)
df.to_csv(r'F:\Eduard\Autodezvoltare\CV camp\CVC5\Gun_detection_xml\gun_detection.csv')
df
```
